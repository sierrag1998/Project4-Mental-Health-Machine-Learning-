```python
#Import Dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
#Import CSV

path = "Data/mental_health2.csv"
df = pd.read_csv(path)
df
```

```python
df.loc[df['value'] <= 30.0, 'Is_High_Presence'] = 0
df.loc[df['value'] >= 30.0, 'Is_High_Presence'] = 1
```

```python
df
```

```python
filtered_df = df[["indicator","group_label","state","subgroup","time_period","Is_High_Presence"]]
filtered_df
```

```python
cat = filtered_df.dtypes[filtered_df.dtypes == "object"].index.tolist()
cat
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
new_df = pd.get_dummies(data=filtered_df, columns=['indicator', 'group_label', 'state', 'subgroup'])
new_df
```

```python
new_df.to_csv('Data/mental_health_cleaned.csv', index=False)
```
